## remove New line chars

```python
from re import sub
Q = sub("[\n]"," ",Q)
Q = sub("[^a-zA-Z0-9 ]","",Q)
```

```python
Q
```

## Unify cases and Tokenizing

```python
Q = Q.upper().split(" ")
```

```python
Q
```

## Convert data in 3-->1 format

```python
inp = []
op = []
for i in range (3,len(Q),1):
    w1 = Q[i-3]
    w2 = Q[i-2]
    w3 = Q[i-1]
    w4 = Q[i]
    inp.append(w1+ " " + w2 + " " + w3)
    op.append(w4)
```

```python
inp
```

## Create Dataframe for the same

```python
import pandas as pd
Q1 = pd.DataFrame([inp,op]).T
```

```python
Q1.columns = ["inp" , "op"]
Q1
```

## Define Array for X

```python
len(inp)
```

```python
from numpy import unique
unique_words = list(unique(Q))
```

```python
unique_words
```

```python
len(unique_words)
```

```python
from numpy import zeros
X = zeros((862,289,3))
```

```python
X
```

## fill data in Blank array X

```python
for line_no , line in enumerate(inp):
    for word_no , word in enumerate(line.split(" ")):
        X[line_no,unique_words.index(word),word_no] = 1
```

## Prepare Output

```python
Y = zeros((862,289))
for line_no, word in enumerate(list(Q1.op)):
    Y[line_no,unique_words.index(word)]=1
```

## Create LSTM model

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,LSTM
from keras.layers.core import Activation
```

```python
X.shape
```

```python
nn = Sequential()
nn.add(LSTM(100,input_shape=(289,3)))
nn.add(Dense(289))
nn.add(Activation("softmax"))
```

```python
nn.compile(loss="categorical_crossentropy",metrics="accuracy",optimizer="adam")
model = nn.fit(X,Y,validation_split=0.2,epochs=30)
```

```python
acc = model.history['accuracy']
val_acc = model.history['val_accuracy']

import matplotlib.pyplot as plt
plt.plot(acc)
plt.plot(val_acc)
```
